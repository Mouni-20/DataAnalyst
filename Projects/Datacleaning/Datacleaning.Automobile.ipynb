6.encode values using sklearn 

```python
from sklearn import datasets
```

```python
#Encoding - To convert any Alphabetical data into numbers
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
Auto['make'] = le.fit_transform(Auto['make'])
Auto['fuel-type'] = le.fit_transform(Auto['fuel-type'])
Auto['aspiration'] = le.fit_transform(Auto['aspiration'])
Auto['num-of-doors'] = le.fit_transform(Auto['num-of-doors'])
Auto['body-style'] = le.fit_transform(Auto['body-style'])
Auto['drive-wheels'] = le.fit_transform(Auto['drive-wheels'])
Auto['engine-location'] = le.fit_transform(Auto['engine-location'])
Auto['engine-type'] = le.fit_transform(Auto['engine-type'])
Auto['num-of-cylinders'] = le.fit_transform(Auto['num-of-cylinders'])
Auto['fuel-system'] = le.fit_transform(Auto['fuel-system'])
print(Auto)
```
