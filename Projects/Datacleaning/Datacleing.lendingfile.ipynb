#1.Remove Null Values

```python
lend.isnull().sum() #Used to check each coloumn has how many null values
```

```python
del lend['Unnamed: 10'] #deleting this whole column helps in removing null value in each row
```

```python
lend.info() # we still see null values in other rows
```

```python
lend = lend.dropna() #using the dropna removes remaining null values in dataframe
```

```python
lend.shape # we can use change in entriess and column of dataframe
```

```python
lend.info() #Using info we can check if all entries are non null or not.
#Finally removed all Null values here
```

```python
lend.isnull().sum()  # 0 Null values in dataframe
```

#2.Replace '-' (dash) with column mean value

```python
lend = lend.replace('-', np.nan) #lets first replace '-' with NaN value
```

```python
lend.info() # we can see - got replaced 
```

```python
lend = lend.fillna(lend.mean()) #lets fill nan with mean 
```

```python
lend.isnull().sum() # check for no Null values
```

```python
lend # finally '-' got replaced column mean values  
```

#3.Encode Lender Asset Size column with 0, 1, 2 (>$50B, $10B-$50B, >$10B)

```python
lend.columns
```

```python
lend['Lender Asset Size'].unique() #callout the specofic column which we need to encode
```

```python
lend['Lender Asset Size'] = lend['Lender Asset Size'].replace(['>$10B               ','$10B-$50B           ','>$50B               '],[0,1,2])
```

```python
lend['Lender Asset Size'].unique() 
```

```python
lend #Encoded the values in Lender Asset Size
```

4.Replace '.'(fullstop) with column mean value

```python
lend['CC Amount/TA1'].unique()
```

```python
lend['CC Amount/TA1'] =lend['CC Amount/TA1'].replace('    .  ', np.nan)
```

```python
lend = lend.fillna(lend.mean())
```

```python
lend.tail() #reolaced . with column  mean
```
